```python
import pandas as pd
from IPython.display import display
from PIL import Image
from matplotlib import pyplot as plt
from os.path import join
```

## Datos originales

```python
BASE_PATH_IMAGES_TRAIN = '../../datasets/covid/train/'
dfTrain = pd.read_csv('../../datasets/covid/train.txt', sep = ' ')
dfTrain.head()
```

```python
def plotImages(img1,img2):
    _, ax = plt.subplots(1, 2, figsize=(20, 20), sharex=True, sharey=True)
    ax[0].imshow(img1)
    ax[0].title.set_text('Positive')

    ax[1].imshow(img2)
    ax[1].title.set_text('Negative')

    plt.show()
```

```python
rstate = 10
positivo = dfTrain[dfTrain['class'] == 'positive'].sample(1,random_state=rstate)
negativo = dfTrain[dfTrain['class'] == 'negative'].sample(1, random_state=rstate)
positivo_img = Image.open(join(BASE_PATH_IMAGES_TRAIN, positivo['filename'].values[0])).convert('RGB')
negativo_img = Image.open(join(BASE_PATH_IMAGES_TRAIN, negativo['filename'].values[0])).convert('RGB')
plotImages(positivo_img, negativo_img)
```
